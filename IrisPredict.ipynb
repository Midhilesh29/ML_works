```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv(r'Iris.csv')
data=pd.read_csv('Iris.csv')
df.drop('Id',axis=1,inplace=True)
data.drop('Id',axis=1,inplace=True)
```

```python
df['Species'].astype('category')
df['target']=pd.Categorical(df['Species']).codes
```

```python
df.head()
```

```python
fea=df.columns.values.tolist()
feature=df[fea[0:4]]
label=df['target']
```

```python
feature.head()
```

```python
import matplotlib.pyplot as plt
pd.scatter_matrix(feature,alpha=0.9,figsize=(10,10))
plt.show()
```

```python
feature.plot(kind='box',subplots=True,layout=(2,2),figsize=(10,10))
```

```python
feature.corr()
```

```python
sb.pairplot(data,hue="Species")
```

```python
x_train,x_test,y_train,y_test=train_test_split(feature,label,test_size=0.33)
```

```python
classifier = LogisticRegression()
classifier.fit(x_train, y_train)

y_pred = classifier.predict(x_test)

# Summary of the predictions made by the classifier
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print("Accuracy score",accuracy_score(y_pred,y_test))
```

```python
classifier = GaussianNB()
classifier.fit(x_train, y_train)

y_pred = classifier.predict(x_test)

# Summary of the predictions made by the classifier
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print("Accuracy score",accuracy_score(y_pred,y_test))
```

```python
from sklearn.svm import SVC

classifier = SVC()
classifier.fit(x_train, y_train)

y_pred = classifier.predict(x_test)

# Summary of the predictions made by the classifier
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print("Accuracy score",accuracy_score(y_pred,y_test))
```

```python
classifier = KNeighborsClassifier(n_neighbors=8)
classifier.fit(x_train, y_train)

y_pred = classifier.predict(x_test)

# Summary of the predictions made by the classifier
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print("Accuracy score",accuracy_score(y_pred,y_test))
```

```python
classifier = DecisionTreeClassifier()

classifier.fit(x_train, y_train)

y_pred = classifier.predict(x_test)

# Summary of the predictions made by the classifier
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print("Accuracy score",accuracy_score(y_pred,y_test))
```
